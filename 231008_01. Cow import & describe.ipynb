```python
import pandas as pd
import numpy as np
import os
```

```python
### 소 데이터 불러오기
cow_22_12 = pd.read_excel("Cow Data/2022-12.xlsx")
cow_23_01 = pd.read_excel("Cow Data/2023-01.xlsx")
cow_23_02 = pd.read_excel("Cow Data/2023-02.xlsx")
cow_23_03 = pd.read_excel("Cow Data/2023-03.xlsx")
cow_23_04 = pd.read_excel("Cow Data/2023-04.xlsx")
cow_23_05 = pd.read_excel("Cow Data/2023-05.xlsx")
cow_23_06 = pd.read_excel("Cow Data/2023-06.xlsx")
cow_23_07 = pd.read_excel("Cow Data/2023-07.xlsx")
cow_23_08 = pd.read_excel("Cow Data/2023-07.xlsx")
```

```python
### 데이터 년/월 변수 만들기
cow_22_12['ym'] = '22_12'
cow_23_01['ym'] = '23_01'
cow_23_02['ym'] = '23_02'
cow_23_03['ym'] = '23_03'
cow_23_04['ym'] = '23_04'
cow_23_05['ym'] = '23_05'
cow_23_06['ym'] = '23_06'
cow_23_07['ym'] = '23_07'
cow_23_08['ym'] = '23_08'
```

```python
### 데이터 셋 합치기
cow = pd.concat([cow_22_12, 
                 cow_23_01, 
                 cow_23_02, 
                 cow_23_03,
                 cow_23_04,
                 cow_23_05,
                 cow_23_06,
                 cow_23_07,
                 cow_23_08])
```

```python
a = cow[cow['농가근내평균가산']!='\\N'].copy()
a.info()
```

```python
a['농가근내평균가산'] = a['농가근내평균가산'].astype('float64')
a['농가근내평균가산'].head()
```
